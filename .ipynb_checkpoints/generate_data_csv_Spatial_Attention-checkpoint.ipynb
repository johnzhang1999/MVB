```python
import json
import csv
import random
import numpy as np
```

```python
data_root = '../data/MVB_train/'
data_json = data_root + 'Info/train.json'
train_csv = data_root + 'Info/train.csv'
test_csv = data_root + 'Info/test.csv'
```

```python
with open(data_json,'r') as json_file:
    data = json.load(json_file)
data = data['image']
data
```

```python
data_map = list(map(lambda d:(d['id'],d['image_path'].split('\\')[-1],d['datatype'],d['image_name'].split('_')[2].split('.')[0]),data))
# data_map
lib = {}
for id,path,type,cam in data_map:
    if not id in lib:
        lib[id] = {'probe':[],'gallery':[]}
    if type == 'p':
        lib[id]['probe'].append((path,cam))
    elif type == 'g':
        lib[id]['gallery'].append((path,cam))
```

```python





random.shuffle(data_map)
RATIO = 0.15
test_len = round(len(data_map) * RATIO)
train = data_map[:-test_len]
test = data_map[-test_len:]
print(len(train),len(test))
print(len(test)/(len(train)+len(test)))

```

```python
with open(train_csv,'w') as csv_file:
    writer = csv.writer(csv_file,delimiter=',')
    for img in train:
        writer.writerow(img)
with open(test_csv,'w') as csv_file:
    writer = csv.writer(csv_file,delimiter=',')
    for img in test:
        writer.writerow(img)
```
