```python
import json
import csv
import random
import numpy as np
```

```python
data_root = '../data/MVB_train/'
data_json = data_root + 'Info/train.json'
train_csv = data_root + 'Info/sa_train.csv'
gallery_csv = data_root + 'Info/sa_gallery.csv'
query_csv = data_root + 'Info/sa_query.csv'
```

```python
with open(data_json,'r') as json_file:
    data = json.load(json_file)
data = data['image']
data
```

```python
data_map = list(map(lambda d:(d['id'],d['image_path'].split('\\')[-1],d['datatype'],d['image_name'].split('_')[2].split('.')[0]),data))
# data_map
lib = {}
for id,path,type,cam in data_map:
    if not id in lib:
        lib[id] = {'probe':[],'gallery':[]}
    if type == 'p':
        lib[id]['probe'].append((path,cam))
    elif type == 'g':
        lib[id]['gallery'].append((path,cam))
lib
```

```python
items = list(lib.items())
random.shuffle(items)
# train_size = round(len(items)*0.8) -> test 800
train_size = len(items) - 500

train_lib = dict(items[:train_size])
test_lib = dict(items[train_size:])

train = []
for pid,ps_and_gs in train_lib.items():
    for p_or_g,imgs in ps_and_gs.items():
        for path,cam in imgs:
            train.append((path,pid,cam))
gallery = []
for pid,ps_and_gs in test_lib.items():
    for path,cam in ps_and_gs['gallery']:
        gallery.append((path,pid,cam))

query = []
for pid,ps_and_gs in test_lib.items():
    for path,cam in ps_and_gs['probe']:
        query.append((path,pid,cam))
```

```python
with open(train_csv,'w') as csv_file:
    writer = csv.writer(csv_file,delimiter=',')
    for img in train:
        writer.writerow(img)
with open(gallery_csv,'w') as csv_file:
    writer = csv.writer(csv_file,delimiter=',')
    for img in gallery:
        writer.writerow(img)
with open(query_csv,'w') as csv_file:
    writer = csv.writer(csv_file,delimiter=',')
    for img in query:
        writer.writerow(img)
```
